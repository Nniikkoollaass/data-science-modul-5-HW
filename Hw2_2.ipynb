```python
import pandas as pd
import requests
from io import StringIO
```

```python
# Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
spreadsheet_url = 'https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/export?format=csv&gid=1724592523'
# Вміст CSV-файлу з Google Sheets
response = requests.get(spreadsheet_url)
csv_data = StringIO(response.content.decode('utf-8'))
df = pd.read_csv(csv_data, encoding='utf-8')
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
# Порахуйте кількість пропусків у кожному стовпці
missing_values = df.isnull().sum()

# Порахуйте загальну кількість значень у кожному стовпці
total_values = df.shape[0]

# Розрахуйте частку пропусків для кожного стовпця
missing_ratio = (missing_values / total_values) * 100

# Виведіть результат
missing_ratio
```

```python
#Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"

# Видаліть усі стовпці, в яких є пропуски, крім "Мова програмування"
columns_to_drop = df.columns[df.isnull().any()].difference(['Мова програмування'])
df_cleaned = df.drop(columns=columns_to_drop)

# Виведіть нову таблицю після видалення пропусків
df_cleaned
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
# Порахуйте кількість пропусків у кожному стовпці
missing_values_after = df_cleaned.isnull().sum()

# Порахуйте загальну кількість значень у кожному стовпці
total_values_after = df_cleaned.shape[0]

# Розрахуйте частку пропусків для кожного стовпця
missing_ratio_after = (missing_values_after / total_values_after) * 100

# Виведіть результат
print(missing_ratio_after)

# Перевірте, які стовпці залишилися
df_cleaned.columns
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
# Видаліть усі рядки з пропусками


### Можна не створювати копій, якщо немає необхідності

# df_some_copy = df_cleaned.copy()
# df_cleaned = df_some_copy.dropna()
df_cleaned = df_cleaned.dropna()

# Виведіть нову таблицю без рядків з пропусками
df_cleaned
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
print(df_cleaned.shape)
```

```python
# Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
# Відфільтруйте рядки, де мова програмування - Python


### Те ж питання, ви повинні продовжувати працювати з вашим датафреймом
# python_data = df[df['Мова програмування'] == 'Python']
python_data = df_cleaned[df_cleaned['Мова програмування'] == 'Python']

# Виведіть нову таблицю зі спеціалістами, які програмують на Python
python_data
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
# Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
# Виконайте групування за стовпцем "Посада"
grouped_data = python_data.groupby('Посада')

# Виведіть перші рядки з кожної групи
for posada, group in grouped_data:
    print(f"Посада: {posada}")
    group.head()  # Виведіть перші рядки з кожної групи
```

```python
# Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
# Виконайте групування за стовпцем "Посада" та агрегацію за допомогою методу agg()
aggregated_data = python_data.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})

# Виведіть результат
aggregated_data
```

# У функції ви повинні рахувати середнє

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    # Отримати значення зарплати на місяць
    # salary = row['Зарплата на місяць']
    salary = row['Зарплата на місяць'].mean()

    # Повернути значення зарплати на місяць
    return salary

# Застосуйте функцію fill_avg_salary до кожного рядка за допомогою методу apply

### Ви рахуєте середнє не серед макс і мін, а серед всіх точок даних і додаєте її до вашого згрупованого датафрейму

# python_data.loc[:, 'avg'] = python_data.apply(fill_avg_salary, axis=1)
aggregated_data[('Зарплата на місяць', 'avg')] = grouped_data.apply(fill_avg_salary)

##### або просто так, якщо б у вас був не подвійний заголовок
##### aggregated_data['avg'] = grouped_data.apply(fill_avg_salary)


# Обчисліть середнє значення зарплати на місяць і створіть новий стовпець "avg"
# avg_salary = python_data['Зарплата на місяць'].mean()

# Заповніть всі рядки нового стовпця "avg" середнім значенням зарплати на місяць
# python_data.loc[:, 'avg'] = avg_salary

# Виведіть таблицю з новим стовпцем "avg"
aggregated_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot 1: Bar plot showing missing values in each column
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_values.index, y=missing_values.values)
plt.title('Missing Values in Each Column')
plt.xlabel('Columns')
plt.ylabel('Missing Values Count')
plt.xticks(rotation=90)
plt.show()

# Plot 2: Pie chart representing the ratio of missing values in each column
plt.figure(figsize=(8, 8))
plt.pie(missing_ratio, labels=missing_ratio.index, autopct='%1.1f%%', startangle=140)
plt.title('Ratio of Missing Values in Each Column')
plt.axis('equal')
plt.show()

# Plot 3: Box plot of salaries by programming language
plt.figure(figsize=(10, 6))
sns.boxplot(x='Мова програмування', y='Зарплата на місяць', data=df)
plt.title('Distribution of Salaries by Programming Language')
plt.xlabel('Programming Language')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()

# Plot 4: Histogram of salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Зарплата на місяць'], bins=20, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```
